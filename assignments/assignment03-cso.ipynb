```python
# Code to retrieve "exchequer account (historical series)" from the CSO, and stores it into a file called "cso.json".

# Packages required
import requests
import json

# URL obtained from cso.ie
url = 'https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en'
response = requests.get(url)
print (response)
# Response code of 200 so its working 
data = response.json()
with open("cso.json", "wt") as fp:
    json.dump(data, fp)
```
