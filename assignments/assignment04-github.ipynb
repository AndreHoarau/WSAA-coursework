This is the location of the repo and to get some content I made a txt file from copying https://en.wikipedia.org/wiki/Andrew_Garfield detailing the life and career of Andrew Garfield, not quite the same as Andrew Beatty, but it gives me multiple instances of the word Andrew to write.

Step 2: Using Github go to developer settings and generate a fine grained token that can only be used to read and write my repository contents.

```python
# Imports that will be needed to do this including requests and the config file with my api_keys
import requests
from config import api_keys as apk
# Ensure to pip install pyGithub
from github import Github 
```

```python
# Create the url variable
url ='https://api.github.com/repos/AndreHoarau/wsaa_testprivaterepo'
# Define our api key
apikey = apk["andrew_txt"]
headers = {
    "Authorization": f"Bearer {apikey}",
    "Accept": "application/vnd.github+json"
}
# Get request of the repo with check of a status
response = requests.get(url, headers = headers)
print (response.status_code)
```

```python
# Authenticate using apikey
g = Github(apikey)
# Get the repository object
repo = g.get_repo("AndreHoarau/wsaa_testprivaterepo")
# Get the txt file and download as raw text
file_info = repo.get_contents("andrew.txt")
content = requests.get(file_info.download_url).text
# Replace Andrew with Andre 
updated_content = content.replace("Andrew", "Andre")
# Update on Github with new content 
repo.update_file(file_info.path, "Replace 'Andrew' with 'Andre'", updated_content, file_info.sha)
# Confirm that it has been done
print("File updated successfully!")
```
